```python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.linear_model import RidgeClassifier, LogisticRegression 
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

clfs = [RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier,
       RidgeClassifier, LogisticRegression,
       LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis, KNeighborsClassifier, SVC]

clfs = map(
    lambda x: str(x()).replace("\n", " ").replace("=",":"), clfs
)

print("clfs = [")
for clf in clfs: 
    parts = clf.split("(")
    clf_name = parts[0]
    params = parts[1].replace(")", "")
    print("(" + clf_name, ",")
    params = list(map(lambda x: x.strip().replace('"', ""), params.split(",") ))
    print(" {")
    for p in params:
        p = p.split(":")
        print("\t'" + p[0] + "': [" + p[1] + "]," )
    print("}),")
print("]")
```
